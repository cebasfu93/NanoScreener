```python
from pathlib import Path

EXP_DATA_PATH = Path("/Users/sebastian/Documents/PhD/Data/ThiolScreening/SEB_3XREPLICA/ExperimentalData/ExpData.csv")
COMP_TEST_DATA_PATH = Path("/Users/sebastian/Documents/PhD/Data/ThiolScreening/SEB_3XREPLICA/TestSet/COMBINED.csv")
COMP_TRAIN_DATA_PATH = Path("/Users/sebastian/Documents/PhD/Data/ThiolScreening/SEB_3XREPLICA/TrainingSet/TRAIN_ALL.csv")
REPLICAS = [1]
T_exp = 273.15 + 25
RT_exp = T_exp * 8.314 / (4.184 * 1000)  # kcal/mol
T_comp = 300
RT_comp = T_comp * 8.314 / (4.184 * 1000)  # kcal/mol
```

```python
import pandas as pd

# the Numbers app saved the CSV with semicolons
# I editted the CSV to include Andreas KA's for the   test set
exp_data = pd.read_csv(EXP_DATA_PATH, skiprows=4, delimiter=";", index_col="Name")  
exp_data.index = pd.Index(map(lambda x: x.upper(), exp_data.index))
exp_data.loc[:, "DG (kcal/mol)"] = exp_data.loc[:, "delta G exp (kJ/mol)"] / 4.184
exp_data.loc[:, "DG std (kcal/mol)"] = exp_data.loc[:, "delta G exp std"] / 4.184
exp_data.drop(labels=["delta G exp (kJ/mol)", "delta G exp std"], axis="columns", inplace=True)
exp_data.head()
```

```python
def import_clean_data(full_path: Path, replicas: list[int]) -> pd.DataFrame:
    rename_col_map = {f"DG{i}": f"W{i}"  for j in replicas for i in range(1 + 10 * (j - 1), 11 + 10 * (j - 1))}
    data = pd.read_csv(full_path, index_col="Names")
    data.index = pd.Index(map(lambda x: x.upper(), data.index))
    data.rename(columns=rename_col_map, inplace=True)
    data = data.loc[:, list(rename_col_map.values())]
    raw_scores_cols = list(rename_col_map.values())
    data.loc[:, "W Q1"] = data.loc[:, raw_scores_cols].quantile(0.25, axis=1)
    data.loc[:, "W Q2"] = data.loc[:, raw_scores_cols].quantile(0.50, axis=1)
    data.loc[:, "W Q3"] = data.loc[:, raw_scores_cols].quantile(0.75, axis=1)
    data.loc[:, "W mean"] = data.loc[:, raw_scores_cols].mean(axis=1)
    data.loc[:, "W std"] = data.loc[:, raw_scores_cols].std(axis=1)
    data.loc[:, "W Q2 - Q1"] = data.loc[:, "W Q2"] - data.loc[:, "W Q1"]
    data.loc[:, "W Q3 - Q2"] = data.loc[:, "W Q3"] - data.loc[:, "W Q2"]
    data.loc[:, "W sem"] = data.loc[:, [f"W{i}" for j in replicas for i in range(1 + 10 * (j - 1), 11 + 10 * (j - 1))]].sem(axis=1)
    return data

train_data = import_clean_data(full_path=COMP_TRAIN_DATA_PATH, replicas=REPLICAS)
test_data = import_clean_data(full_path=COMP_TEST_DATA_PATH, replicas=REPLICAS)
print(f"Training data {train_data.shape}")
display(train_data.head())
print(f"Test data {test_data.shape}")
display(test_data.head())
```

```python
from typing import Literal


def select_exp_data(exp_data: pd.DataFrame, selection: Literal["Train", "Test"], only_active: bool = True) -> pd.DataFrame:
    selected_data = exp_data.loc[exp_data.loc[:, "Training/Test"] == selection, :]
    if only_active is True:
        return selected_data.loc[selected_data.loc[:, "DG (kcal/mol)"] != 0.0, :]
    return selected_data
```

```python
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np 

train_col = "W Q2"
train_col_err = ["W sem"] #["W Q2 - Q1", "W Q3 - Q2"]
y = exp_train_active.loc[:, "DG (kcal/mol)"]
y_err = exp_train_active.loc[:, "DG std (kcal/mol)"]
x = train_data.loc[y.index, train_col]
x_err = train_data.loc[y.index, train_col_err].T

x_range_min = -30
y_range_min = -10
slope, intercept , r_value, *_ = linregress(x, y)
space = np.linspace(0, x_range_min, 3)
print(f"y = {slope:.2f}x + {intercept:.2f}, R2 = {r_value**2:.2f}")

FONTSIZE = 14
BLUE = (0.0, 0.4, 0.8)
fig, ax = plt.subplots(figsize=(6, 4))
ax.errorbar(x=x, y=y, xerr=x_err, yerr=y_err, fmt="o", color=BLUE, ecolor="k", elinewidth=1, capsize=3, mew=1, mec="k")
ax.plot(space, intercept + slope * space, color="k", linestyle="--", label=f"y = {slope:.2f}x - {abs(intercept):.2f}"+ "\n" +  fr"$R^2$ = {r_value**2:.2f}")
ax.tick_params(labelsize=FONTSIZE, direction="in", length=5, width=1)
ax.set_xlabel(r"$W$ ($kcal$ $mol^{-1}$)", fontsize=FONTSIZE)
ax.set_ylabel(r"$\Delta G_{exp}$ ($kcal$ $mol^{-1}$)", fontsize=FONTSIZE)
ax.set_xlim(x_range_min, 0)
ax.set_ylim(y_range_min)
ax.legend(loc="lower left", fontsize=FONTSIZE - 2, bbox_to_anchor=[0.55,0.03,1,0.2])
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

ax2 = plt.axes([0.23, 0.65, 0.3, 0.25])
ax2.errorbar(x=range(len(x)), y=y/x, fmt="o", ms=4, color=BLUE)
ax2.set_xlim(-1, len(x) + 1)
ax2.set_xlabel("Dyad ID", fontsize=FONTSIZE - 2)
ax2.set_ylabel(r"$\Delta G_{exp}$/ $W$", fontsize=FONTSIZE - 2)
ax2.tick_params(labelsize=FONTSIZE - 2, direction="in", length=5, width=1)
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])
ax2.axhline((y/x).mean(), linestyle="-", color="gray", lw=1)
ratio = (y/x).mean()
print(f"G_exp/W: {ratio:.2f}")
plt.savefig("training_set_fit.png", format="png", dpi=300, bbox_inches="tight")
plt.savefig("training_set_fit.svg", format="svg", dpi=300, bbox_inches="tight")
plt.show()
plt.close()
```
