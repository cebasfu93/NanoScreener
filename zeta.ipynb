```python
from pathlib import Path

ZETA_PATH = Path("/Users/sebastian/Documents/PhD/Data/ThiolScreening/ZP.xlsx")
```

```python
import pandas as pd

zeta = pd.read_excel(ZETA_PATH, )
zeta
```

```python
import numpy as np
from matplotlib import pyplot as plt

FONTSIZE = 14
BLUE = (0.0, 0.4, 0.8)
fig, ax = plt.subplots(figsize=(5, 3))

ax.errorbar(
    x=zeta.loc[:, "Rank"], y=zeta.loc[:, "ZP"], yerr=zeta.loc[:, "std.dev"], 
    fmt="o", color=BLUE, mec="k", mew=0.5, ecolor="k", capsize=5, ms=6, lw=0.5
)
ax.set_ylim(0, -60)
ax.set_xlabel("Ranking position", fontsize=FONTSIZE)
ax.set_ylabel("Zeta potential (mV)", fontsize=FONTSIZE)
ax.tick_params(labelsize=FONTSIZE)
ax.set_yticks(np.linspace(0, -60, 7))
ax.set_xticks(np.linspace(0, 100, 6))
ax.set_xlim(0, 100)
plt.savefig("zeta_potential.png", format="png", dpi=300, bbox_inches="tight")
plt.savefig("zeta_potential.svg", format="svg", dpi=300, bbox_inches="tight")
plt.show()
plt.close()
```
